```python
def freed_prisoners(prison):
    if prison[0] == 0:
        return 0
    
    freed = 0
    target = 1 # Initially looking for unlocked cells
    
    for cell in prison:
        if cell == target:
            freed += 1
            # Flip the target: 1 becomes 0, 0 becomes 1
            target = 1 - target                  
    return freed
```

```python
# Test 1
freed_prisoners(([1, 1, 0, 0, 0, 1, 0])) # Output: 4
```

```python
# Test 2
freed_prisoners(([1, 1, 1])) # Output: 1
```

```python
# Test 3
freed_prisoners(([0, 0, 0])) # Output: 0
```

```python
# Test 4
freed_prisoners(([0, 1, 1, 1])) # Output: 0
```
