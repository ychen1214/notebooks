```python
def help_bobby(size):      
    row=0    
    list =[]    
    t=False
    if size<0:
        size = abs(size)
        t=True
    for column in range(size):
        array=[[0]*size]
        p = column
        if t:
            p = size-1-column
        array[row][p]=1        
        list.append(array)                
    return list
```

```python
help_bobby(0)
```
