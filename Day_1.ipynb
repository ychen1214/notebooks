```python
#Video Streaming Data Analysis
import pandas as pd

data = {'video_id': ['1', '2', '3', '4'],
        'title': ['Amazing Adventure', 'Mystery of the Deep', 'City Lights', 'The Great Escape'],
        'genre': ['Action', 'Mystery', 'Drama', 'Action'],
        'release_year': [2020, 2021, 2022, 2023],
        'duration': [120, 140, 115, 134],
        'view_count': [2500000, 500000, 1500000, 3500000],
        }
video_stream = pd.DataFrame(data)

video_stream.fillna(0, inplace=True)
video_stream['release_year'] = video_stream['release_year'].astype(int)
video_stream['duration'] = video_stream['duration'].astype(int)
video_stream['view_count'] = video_stream['view_count'].astype(int)

video_stream[(video_stream['release_year'] > 2021) & (video_stream['view_count'] > 1000000)]
```

```python
#CRM Summary Report
import pandas as pd

crm_data = {'customer_id': [1, 2, 3, 4],
            'name': ['Alice', 'Bob', 'Charlie', 'David'],   
            'email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'david@example.com'],
}
crm_df = pd.DataFrame(crm_data)

crm_orders_data = {'customer_id': [1, 2, 1, 3],
                    'order_id': [101, 102, 103, 104],
                    'amount': [100.0, 200.0, 150.0, 300.0],
                    'order_date': ['2023-01-15', '2023-02-20', '2023-03-10', '2023-04-05'],
                    'product_id': [1, 2, 3, 1],
                   }
crm_orders_df = pd.DataFrame(crm_orders_data)

crm_products_data = {'product_id': [1, 2, 3],
                     'product_name': ['Product A', 'Product B', 'Product C'],   
                     'category': ['Category 1', 'Category 2', 'Category 1'],

}
crm_products_df = pd.DataFrame(crm_products_data)

df_view = pd.merge(crm_df, crm_orders_df, on='customer_id', how='inner')
df_view = pd.merge(df_view, crm_products_df, on='product_id', how='inner')
df_view['order_date'] = pd.to_datetime(df_view['order_date']) 

df_view[['order_date','name','email','product_name','category','amount']]
```

```python
#Social Meida PII Extraction
import pandas as pd

social_media_data = {
                     'user_id': [101, 102, 103, 104],   
                     'email': ['user1@example.com', 'user2@example.com', 'user3@example.com', 'user4@example.com'],
                     'phone': ['555-1234', '555-5678', '555-8765', '555-4321'],
                    }
social_media_df = pd.DataFrame(social_media_data)
social_media_df.fillna(0, inplace=True)
social_media_df['email'] = social_media_df['email'].str.replace(r'(?<=.)[^@]+(?=@)',r'***', regex=True)
social_media_df['phone'] = social_media_df['phone'].str.replace(r'(\d{3})-(\d{4})',r'***-\2', regex=True)
social_media_df
```

```python
#Manufacturing Defect Rate Analysis 
import pandas as pd

manufacturing_data = {'product_id': [1, 2, 3, 4],
                      'category': ['A', 'B', 'A', 'B'],
                      'product_name': ['Product A', 'Product B', 'Product C', 'Product D'],
                      'defect': [0, 1, 0, 1]    
}
manufacturing_df = pd.DataFrame(manufacturing_data)

manufacturing_sales_data = {
                           'sales_id': [1, 2, 3, 4],
                           'product_id': [1, 2, 3, 4],
                           'quantity': [1000, 1500, 1200, 800],
                           'revenue': [10000, 15000, 12000, 8000]
}
manufacturing_sales_df = pd.DataFrame(manufacturing_sales_data)

df_view = pd.merge(manufacturing_df, manufacturing_sales_df, on='product_id', how='inner')
df_view.fillna(0, inplace=True)
df_view = df_view.sort_values(by=['category', 'revenue'], ascending=False)
# +1 makes it 1-indexed like SQL
df_view['row_num'] = df_view.groupby('category').cumcount() + 1
df_view[['category', 'product_name', 'revenue', 'row_num']]
```

```python
#Call Center Performance Metrics
import pandas as pd

calls_data = {'call_id': [1, 2, 3, 4],
              'cust_id': [101, 102, 101, 103],
              'duration': [120, 180, 90, 210],
              'date': ['2023-01-01', '2023-01-01', '2023-01-03', '2023-01-04']
              }  
calls_df = pd.DataFrame(calls_data)
calls_df.fillna(0, inplace=True)

customer_data = {'cust_id': [101, 102, 103],
                 'name': ['Alice', 'Bob', 'Charlie'],   
                 'email': ['alice@example.com', 'bob@example.com', 'charlie@example.com'],
                 'state': ['NY', 'CA', 'IL'],
                 'tenure': [12, 24, 6],
                 'occupation': ['Engineer', 'Teacher', 'Student']
                }
customer_df = pd.DataFrame(customer_data)
customer_df.fillna(0, inplace=True)

df_view = pd.merge(calls_df, customer_df, on='cust_id', how='inner')
df_view['date'] = pd.to_datetime(df_view['date'])
df_view.groupby('date').agg(
    total_calls=('call_id', 'count'),
    avg_duration=('duration', 'sum')
)
```

```python
#Research Paper Citation Analysis
import pandas as pd

ai_authors_data = {
    'paper_id': [1, 1, 2, 3],
    'author_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
}
ai_authors_df = pd.DataFrame(ai_authors_data)
ai_authors_df.fillna(0, inplace=True)

ai_research_papers_data = {
    'paper_id': [1, 2, 3],
    'title': ['AI in Healthcare', 'Machine Learning Basics', 'Deep Learning Advances'],
    'publication_year': [2020, 2021, 2022]
}
ai_research_papers_df = pd.DataFrame(ai_research_papers_data)
ai_research_papers_df.fillna(0, inplace=True)

df_view = pd.merge(ai_authors_df, ai_research_papers_df, on='paper_id', how='inner')
df_view['row_number'] = df_view.groupby('paper_id').cumcount() + 1
df_view[['author_id','name','paper_id','row_number']]
```

```python
#F&B Product Sales Summary
fnb_inventory_data = {
    'product_id': [1, 2, 3, 4],
    'product_name': ['Coffee', 'Tea', 'Juice', 'Water'],    
    'price': [3.50, 2.50, 4.00, 1.50],
    'stock': [100, 150, 80, 200],
    'warehouse_location': ['A1', 'A2', 'B1', 'B2']
}
fnb_inventory_df = pd.DataFrame(fnb_inventory_data) 
fnb_inventory_df.fillna(0, inplace=True)

fnb_products_data = {
    'product_id': [1, 2, 3, 4],
    'name': ['Coffee', 'Tea', 'Juice', 'Water'],
    'category': ['Beverage', 'Beverage', 'Beverage', 'Beverage'],
    'supplier': ['Supplier A', 'Supplier B', 'Supplier C', 'Supplier D']
}
fnb_products_df = pd.DataFrame(fnb_products_data)
fnb_products_df.fillna(0, inplace=True)

fnb_sales_data = {
    'sale_id': [1, 2, 3, 4],
    'product_id': [1, 2, 3, 4],
    'quantity_sold': [30, 50, 20, 70],
    'sale_date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    'revenue': [105.0, 125.0, 80.0, 105.0]
}
fnb_sales_df = pd.DataFrame(fnb_sales_data)
fnb_sales_df.fillna(0, inplace=True)

df_views = pd.merge(fnb_inventory_df, fnb_products_df, on='product_id', how='inner')
df_views = pd.merge(df_views, fnb_sales_df, on='product_id', how='inner')
df_views['sale_date'] = pd.to_datetime(df_views['sale_date'])
df_views.groupby(['product_id', 'name', 'category']).agg(
    total_quantity_sold=('quantity_sold', 'sum'),
    total_revenue=('revenue', 'sum'),
    total_stock=('stock', 'sum')
)

```

```python
#Payroll Calculation Analysis
import pandas as pd

rp_employee_data = {
    'employee_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [30, 40, 35, 28],
    'department': ['HR', 'Engineering', 'Marketing', 'Sales'],
    'position': ['Manager', 'Developer', 'Analyst', 'Salesperson'],
    'salary': [60000, 80000, 70000, 50000]
}
rp_employee_df = pd.DataFrame(rp_employee_data)
rp_employee_df.fillna(0, inplace=True)

rp_paypayroll_data = {
    'employee_id': [1, 2, 3, 4],
    'hours_wokred': [160, 170, 150, 180],
    'hourly_rate': [37.5, 47.06, 46.67, 27.78],
}
rp_paypayroll_df = pd.DataFrame(rp_paypayroll_data)
rp_paypayroll_df.fillna(0, inplace=True)

df_view = pd.merge(rp_employee_df, rp_paypayroll_df, on='employee_id', how='inner')
df_view['total_pay'] = df_view['hours_wokred'] * df_view['hourly_rate']
df_view[['employee_id', 'name', 'department', 'position', 'total_pay']]
```
