```python
#Recursion: Happy Number
def is_happy(n):
    if n == 1:
        return True
    elif n == 4:
        return False
    else:
        sum = 0
        while n > 0:
            digit = n % 10
            sum += digit ** 2
            n //= 10
            print("Test {0},{1},{2}".format(digit, sum, n))
        return is_happy(sum)
```

```python
def is_happy_set(n):
    seen = set()  # Keeps track of numbers we've already visited
    
    while n != 1 and n not in seen:
        seen.add(n)
        
        # Calculate the sum of squares of the digits
        sum_of_squares = 0
        while n > 0:
            digit = n % 10
            sum_of_squares += digit ** 2
            n //= 10
            
        n = sum_of_squares  # Move to the next number in the sequence
        
    return n == 1  # If we stopped because n is 1, it's a happy number
```

```python
is_happy_set(67)
```
