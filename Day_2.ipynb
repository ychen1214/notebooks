```python
#Mining Output Analysis
import pandas as pd

mc_extraction_data = {
    'mine_id': [1, 2, 3, 4],
    'date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04'],
    'mineral': ['Gold', 'Silver', 'Copper', 'Iron'],
    'quantity': [100, 200, 150, 300]
}
mc_extraction_df = pd.DataFrame(mc_extraction_data)
mc_extraction_df.fillna(0, inplace=True)  # Handle missing values if any

mc_mines_data = {
    'mine_id': [1, 2, 3, 4],
    'mine_name': ['Mine A', 'Mine B', 'Mine C', 'Mine D'],
    'location': ['Location A', 'Location B', 'Location C', 'Location D']
}
mc_mines_df = pd.DataFrame(mc_mines_data)
mc_mines_df.fillna('Unknown', inplace=True)  # Handle missing values if any

df_view = pd.merge(mc_extraction_df, mc_mines_df, on='mine_id', how='inner')

df_view.groupby(['mineral', 'location'])['quantity'].sum().reset_index().sort_values(by='location', ascending=True)

```

```python
#E-Commerce Date Dimension Analysis
import numpy as np
import pandas as pd
ecom_dt_orders_data = {
    'order_id': [1, 2, 3, 4],
    'product_id': [101, 102, 103, 104],
    'user_id': [1001, 1002, 1003, 1004],
    'order_date': ['2024-01-01', '2024-01-02', '2024-01-03', '01-02-2024']
}
ecom_dt_orders_df = pd.DataFrame(ecom_dt_orders_data)
ecom_dt_orders_df.fillna(0, inplace=True)  # Handle missing values if any

ecom_dt_products_data = {
    'product_id': [101, 102, 103, 104],
    'product_name': ['Product A', 'Product B', 'Product C', 'Product D'],
    'category': ['Category A', 'Category B', 'Category C', 'Category D']
}
ecom_dt_products_df = pd.DataFrame(ecom_dt_products_data)
ecom_dt_products_df.fillna('Unknown', inplace=True)  # Handle missing values if any

df_view = pd.merge(ecom_dt_orders_df, ecom_dt_products_df, on='product_id', how='inner')
df_view['order_date'] = pd.to_datetime(df_view['order_date'], format='mixed', errors='coerce')  # Convert to datetime, handle errors

result = df_view.assign(
    day_flag = np.where(df_view['order_date'].dt.day > 5, 1, 0)
)[['category', 'order_date', 'product_name', 'user_id', 'day_flag']]
result.sort_values(by='order_date', ascending=True)

```

```python
#Discount Calculation Error Detection
import pandas as pd
import numpy as np

cgdc_sales_data ={
    'storeid':['S101','S102','S103','S104'],
    'productname':['Biscuits','Shampoo','Banana','Tootpaste'],
    'category':['Food','Hygiene','Food','Hygiene'],
    'soldunits':[120,85,150,300],
    'description':['Tasty Biscuits [10% off]','Smoothens Hair [5% off]', 'Fresh Bananas', 'Protects Teeth'],
    'discount': [0.1,0.05,0.0,0.0]
}

cgdc_sales_df = pd.DataFrame(cgdc_sales_data)
cgdc_sales_df['extractdiscount'] =  cgdc_sales_df['description'].str.extract(r'(\d+\.?\d*)%').astype(float) / 100
cgdc_sales_df.fillna(0,inplace=True)
cgdc_sales_df.loc[cgdc_sales_df['productname']=='Tootpaste','extractdiscount']=5
cgdc_sales_df.query('discount != extractdiscount')
```

```python
#Busiest Flight Route Analysis
import pandas as pd

ba_airports_data = {
    'airport_id':['A1','B1','A2','B2','A3','B3'],
    'airport_name':['San Francisco','Los Angeles','New York','Boston','Atlanta','Dallas'],
}
ba_airports_df = pd.DataFrame(ba_airports_data)
ba_airports_df.fillna(0,inplace=True)

ba_flights_data = {
    'flight_id':[1,2,3],
    'origin_airport':['A1','A2','A3'],
    'destination_airport':['B1','B2', 'B3']
}
ba_flights_df = pd.DataFrame(ba_flights_data)
ba_flights_df.fillna(0,inplace=True)

ba_planes_data = {
    'plane_id':[1,2,3],
    'plane_model':['Airbus A320','Boeing 737','Airbus A380']    
}
ba_planes_df = pd.DataFrame(ba_planes_data)
ba_planes_df.fillna(0,inplace=True)
df_view = pd.DataFrame()

df_view = pd.merge(ba_planes_df,ba_flights_df, left_on='plane_id', right_on='flight_id', how='left')
df_view = df_view.merge(ba_airports_df[['airport_id','airport_name']], left_on='origin_airport',right_on='airport_id',how='left')
df_view = df_view.merge(ba_airports_df[['airport_id','airport_name']], left_on='destination_airport',right_on='airport_id',how='left')

df_view = df_view.rename(columns={'airport_name_x': 'origin_airport_str'})
df_view = df_view.rename(columns={'airport_name_y': 'destination_airport_str'})

df_view2 = df_view[['flight_id','origin_airport_str','destination_airport_str','plane_model']]


for col in ['origin_airport_str','destination_airport_str','plane_model']:
    df_view2[f'{col}_len'] = df_view[col].astype(str).map(len)

df_view2[['flight_id','origin_airport_str_len','destination_airport_str_len','plane_model_len']]    
```

```python
#Country GDP Growth Rate

gdp_df1 = pd.DataFrame(
    {
        'country':['USA','USA','China'],
        'year':[2018,2019,2018],
        'gdp':[20544.34,21427.7,13894.04]
    }
)
gdp_df1.fillna(0,inplace=True)

gdp_df2 = pd.DataFrame(
    {
        'country':['China','India','India'],
        'year':[2019,2018,2019],
        'gdp':[14402.72,2713.61,2868.93]
    }
)
gdp_df2.fillna(0,inplace=True)


df_view = pd.concat([gdp_df1,gdp_df2])
df_view = df_view.sort_values(by=['country','year'])

df_view['gdp_growth_rate'] = df_view.groupby('country')['gdp'].pct_change() * 100
df_view 


```
