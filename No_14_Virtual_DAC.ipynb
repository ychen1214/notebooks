```python
#Test Virtual DAC
def dac_to_voltage(decimal_value):
    # Constants for a 10-bit DAC with 5V reference
    resolution_bits = 10
    v_ref = 5.00
    
    # Calculate maximum digital value (1023)
    max_digital_value = (2 ** resolution_bits) - 1
    
    # Ensure input is within the 0-1023 range
    if decimal_value < 0:
        return 0.0
    if decimal_value > max_digital_value:
        return v_ref
        
    # Calculate output voltage
    voltage_out = (decimal_value / max_digital_value) * v_ref
    
    return round(voltage_out, 2)


# Test the function with some values
dac_to_voltage(0)        # Expected output: 0.0
dac_to_voltage(400)      # Expected output: 2.5
dac_to_voltage(1023)     # Expected output: 5.0
```
