```python
# Examples of reading through a dictionary
movie_years = {"The Matrix": 1999, "Se7en": 1995, "The Wizard of Oz": 1939}

# 1. Keys only
for movie in movie_years:
    print(f"Key: {movie}")

# 2. Values only
for year in movie_years.values():
    print(f"Value: {year}")

# 3. Both keys and values
for movie, year in movie_years.items():
    print(f"{movie} was released in {year}")
```
