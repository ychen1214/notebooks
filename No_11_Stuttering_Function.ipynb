```python
#Stuttering function to cause stuttering in the output
def stutter(word):    
    if len(word) > 2:
        return word[0:2] + '...' + word[0:2] + '...' + word
```

```python
#Test 1
stutter('incredible')
```

```python
#Test 2
stutter('enthusiastic')
```

```python
#Test 3
stutter('outstanding')
```
