```python
def reverse_array(arr):
    start = 0
    end = len(arr) - 1    
    while start < end:
        arr[start] = arr[end] 
        arr[end] = arr[start]
        start = start + 1 
        end = end - 1
        
my_array = [1, 2, 3, 4, 5]
print("Original Array:", my_array)
reverse_array(my_array)
print("Reversed Array (in-place):", my_array)
#my_array = [1, 2, 3, 4, 5]
#print("Reversed Array (using reversed()):", list(reversed(my_array)))
```

```python
def has_duplicates(arr):
    n = len(arr)
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] == arr[j]:
                return True
    return False

arr = [1, 2, 3, 4, 3]
print(has_duplicates(arr))  # Output: True
```

```python
def has_duplicates_sorted(arr):
    arr.sort()
    n = len(arr)
    for i in range(n - 1):
        if arr[i] == arr[i+1]:
            return True
    return False

arr = [1, 2, 3, 4, 3]
print(has_duplicates_sorted(arr))  # Output: True
```

```python
def has_duplicates(arr):
    # If the set is smaller than the list, a duplicate was removed
    return len(set(arr)) != len(arr)

arr = [1, 2, 3, 4,3,4]
print(has_duplicates(arr))  # Output: False
```

```python
def binary_search(arr, target):
    arr.sort()  # Ensure the array is sorted
    low = 0
    high = len(arr) - 1

    while low <= high:
        # Check the middle element
        mid = (low + high) // 2
        guess = arr[mid]

        if guess == target:
            return mid  # Success! Found the index
        if guess > target:
            high = mid - 1  # Too high, look left
        else:
            low = mid + 1   # Too low, look right

    return -1  # Target not in list

# Test it out
ordered_numbers = [10, 20, 30, 40, 50]
print(binary_search(ordered_numbers, 40)) # Output: 3
print(binary_search(ordered_numbers, 10)) # Output: 0
```

```python
def compress_string(input_string):
    # Initialize
    current_char = input_string[0]
    char_count = 1
    output = current_char

    # Iterate through the string
    for char in input_string[1:]:
        # If the character matches the current one, increment count
        if char == current_char:
            char_count += 1
        else:  # Append the count to the output and reset for the new character
            output += str(char_count) + char
            current_char = char
            char_count = 1

    # Append the last character's count
    output += str(char_count)

    # If the compressed string is shorter than the original string, return it
    return output if len(output) < len(input_string) else input_string

compress_string("aaabbcaaa")  # Output: "a3b2c1a3"
```

```python
import heapq

def find_kth_largest_heap(nums, k):
    # nlargest returns a list of k largest elements; the last one is the kth
    return heapq.nlargest(k, nums)[-1]

# Example
nums = [3, 2, 3, 1, 2, 4, 5, 5, 6]
print(find_kth_largest_heap(nums, 5))  # Output: 4
```

```python
def fizz_buzz(n):
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

fizz_buzz(15)
```

```python
def fizz_buzz_clean(n):
    for i in range(1, n + 1):
        output = ""
        if i % 3 == 0: output += "Fizz"
        if i % 5 == 0: output += "Buzz"
        print(output or i)

fizz_buzz_clean(15)        
```
