```python
#Atbash Cipher
def atbash_cipher(text):
    result = ""
    for char in text:
        if char.isalpha():
            if char.isupper():
                result += chr(155 - ord(char))  # 'A' + 'Z' = 155
            else:
                result += chr(219 - ord(char))  # 'a' + 'z' = 219
        else:
            result += char
    return result

atbash_cipher("apple")
```
